```python
import pandas as pd
import numpy as np
data = pd.read_excel('data.xlsx')
```

```python
data.head()
```

```python
score_bins = np.arange(0, 100, 10)
```

```python
score_bins = np.arange(0, 100, 10)
def map_bin(x, bins):
    kwargs = {}
    if x == max(bins):
        kwargs['right'] = True
    bin = bins[np.digitize([x], bins, **kwargs)[0]]
    bin_lower = bins[np.digitize([x], bins, **kwargs)[0]-1]
    return '[{0}-{1}]'.format(bin_lower, bin)
```

```python
map_bin(10,score_bins)
```

```python
data = data.dropna()
```

```python
data['Binned_S'] = data['mg_black_score'].apply(map_bin, bins=score_bins)
```

```python
data.head()
```

```python
grouped = data.groupby('Binned_S')
grouped_data = grouped.mean()
```

```python
grouped_data.head(10) 
```

```python
step = score_bins[1]-score_bins[0]
new_index = ['[{0}-{1}]'.format(x, x+step) for x in score_bins]
new_index.pop(-1)
```

```python
grouped_data = grouped_data.reindex(new_index)
```
